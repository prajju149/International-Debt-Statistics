# Part 6: 6 Outliers Transformations

```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("consistent_data.csv")

# Log transform of Total Debt
df['Log Debt'] = np.log1p(df['Total Debt'])

# Boxplot
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['Total Debt'])
plt.title('Boxplot of Total Debt')
plt.tight_layout()
plt.savefig("boxplot_total_debt.png")

df.to_csv("transformed_data.csv", index=False)
```
